```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("setup completed")
```

```python
flight_filepath = r"D:\machine learning project\flight_delays.csv"
flight_data = pd.read_csv(flight_filepath, index_col = "Month")
flight_data
```

```python
#4set width and height of figure
plt.figure(figsize=(10,6))
# set the title of graph
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")
sns.barplot(x=flight_data.index, y = flight_data["AA"],hue = flight_data.index,palette = "viridis")
plt.ylabel("Arrival delay (in minutes)")
plt.show()
```

```python
plt.figure(figsize=(14,6))
plt.title("Average Arrival Delay for Each Airline, by Month")
sns.heatmap(data = flight_data,annot = True)
plt.xlabel("Airline")
plt.show()
```
