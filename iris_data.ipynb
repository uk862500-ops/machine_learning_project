```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("setup completed")
```

```python
iris_filepath = "D:\machine learning project\iris.csv"
iris_data = pd.read_csv(iris_filepath)
iris_data
```

```python
sns.histplot(iris_data["Petal Length (cm)"])
plt.show()
```

```python
sns.kdeplot(data = iris_data["Petal Length (cm)"], shade = True)
plt.show()
```

```python
sns.jointplot(x = "Petal Length (cm)", y = "Sepal Width (cm)", kind = "kde", data = iris_data)
plt.show()
```

```python
sns.histplot(x = "Petal Length (cm)", hue = "Species", data = iris_data)
plt.title("Histogram of Petal Length , by Species")
plt.show()
```

```python
sns.kdeplot(data = iris_data, x = "Petal Length (cm)", hue = "Species", shade = True)
plt.title("Distribution of Petal Length, by Species")
plt.show()
```
