```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
spotify_filepath = r'D:\machine learning project\spotify.csv'
spotify_data = pd.read_csv(spotify_filepath)
spotify_data.head()
```

```python
spotify_data.describe()
```

```python
spotify_data = pd.read_csv(spotify_filepath, index_col = "Date", parse_dates = True)
spotify_data.head()
```

```python
spotify_data.tail()
```

```python
sns.lineplot(data = spotify_data)
```

```python
# Use sns.lineplot instead of just lineplot
plt.figure(figsize=(16,8)) # Makes the chart wider and easier to read
sns.lineplot(data=spotify_data)

# Add a title so you know what you're looking at
plt.title("Daily Streams of Popular Songs on Spotify")
plt.show()
```
